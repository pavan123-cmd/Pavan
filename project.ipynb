```python
bins = [0,12,19,59,100]
labels = ['child','teen','adult','senior']
```

```python
df['AgeGroup'] = pd.cut(df['Age'],bins = bins,labels = labels)
```

```python
sns.barplot(data=df,x='AgeGroup',y='Survived',hue='Pclass',palette='Purples',capsize=0.05,edgecolor='black')
plt.ylim(0,1.0)
plt.title('Survival rate by age group and class')
plt.xlabel('age group')
plt.ylabel('survival probability')
plt.show()
```

```python
sex = df.groupby('Sex',observed=False)['Survived'].mean()
plt.ylim(0,1.0)
sex.plot(kind='bar',color = ['skyblue','salmon'])
plt.title('survival by sex')
plt.xlabel('sex')
plt.ylabel('survial rate')
plt.xticks(rotation=0)
plt.show()
```
